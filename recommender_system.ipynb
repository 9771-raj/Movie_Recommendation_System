```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
%matplotlib inline
```

```python
df=pd.read_csv('movie_database/tmdb_5000_movies.csv')
df2=pd.read_csv('movie_database/tmdb_5000_credits.csv')
movies=df2[['movie_id','cast','crew']]
df_update=df[['title','genres','overview','keywords']]
movies=df_update.join(movies)
```

```python
movies.head()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies['genres'][0]
```

```python
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
#['Action','Adventure','Fantasy','Science Fiction']
```

```python
import ast   # for making string of list to list
ast.literal_eval(movies['genres'][0])
```

```python
def make_list(data):
    list=[]
    for i in ast.literal_eval(data):
        list.append(i['name'])
    return list
```

```python
movies['genres'].apply(make_list)   # apply make list function
```

```python
movies['genres']=movies['genres'].apply(make_list)
```

```python
# similarly we apply make_list function on keywords column
movies['keywords']=movies['keywords'].apply(make_list)
movies.head()
```

```python
def make_list_cast(data):
    list=[]
    count=0
    for i in ast.literal_eval(data):
        if count!=3:
            list.append(i['name'])
            count+=1
        else:
            break
    return list
```

```python
movies['cast'].apply(make_list_cast)
```

```python
movies['cast']=movies['cast'].apply(make_list_cast)
```

```python
def fetch_director(data):
    list=[]
    for i in ast.literal_eval(data):
        if i['job']=='Director':
            list.append(i['name'])
            break
    return list
```

```python
movies['crew'].apply(fetch_director)
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
# now we make list of overview column by using split function
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['genres']+movies['overview']+movies['keywords']+movies['cast']+movies['crew']
movies.head()
```

```python
movie_db=movies[['movie_id','title','tags']]
movie_db.head()
```

```python
movie_db['tags']=movie_db['tags'].apply(lambda x:" ".join(x))
```

```python

movie_db['tags'][0]
```

```python
movie_db['tags']=movie_db['tags'].apply(lambda x:x.lower())
```

```python
movie_db['tags'][0]
```

```python
movie_db.head()
```

```python
# stemming 
import nltk
from nltk.stem import PorterStemmer

ps=PorterStemmer()
def stem(data):
    list=[]
    for i in data.split():
        list.append(ps.stem(i))
    return " ".join(list)
```

```python
movie_db['tags']=movie_db['tags'].apply(stem)
movie_db['tags'][0]
```

```python
# logical part
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
# max_features ---> words in column
# wrt row movie all movie in vector form totalmovie=4800
vectors=cv.fit_transform(movie_db['tags']).toarray()
```

```python
vectors.shape  #sparse matrix
```

use cosine similarity on movie vectors with other movies and itself 
it gives idea about how similarity the movie 
we get a matrix of similarity (4800,4800)

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
similarity
```

```python
similarity.shape
```

do sorting and store index of movies

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=movie_db[movie_db['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(movie_db.iloc[i[0]].title)
```

```python
recommend('John Carter')
```
